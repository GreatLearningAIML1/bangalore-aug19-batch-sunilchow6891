### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
q1 = HR_df['Age'].quantile(.25)
q3 = HR_df['Age'].quantile(.75)
mask = HR_df['Age'].between(q1, q3, inclusive=True)
IQR = HR_df.loc[mask, 'Age']
print(IQR)
```
